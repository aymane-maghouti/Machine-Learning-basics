# Séparation des variables explicatives et la variable à prédire


```python
x1 = data[['X1']]
x2 = data[['X2']]
y = data['Y']
```

# Création du modèle de régression linéaire multiple


```python
X = np.column_stack((x1, x2))
model = LinearRegression().fit(X, y)
```

# Préparation des données pour le tracage de la surface de régression (Hyperplan)

```python
x1_min, x1_max = x1.min(), x1.max()
x2_min, x2_max = x2.min(), x2.max()
x1_range = np.linspace(x1_min, x1_max, 10)
x2_range = np.linspace(x2_min, x2_max, 10)
X1, X2 = np.meshgrid(x1_range, x2_range)
Y = model.predict(np.column_stack((X1.ravel(), X2.ravel()))).reshape(X1.shape)
```

# Tracage de la surface de régression


```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, y)
ax.plot_surface(X1, X2, Y, alpha=0.5)
```

# Tracage des graphiques de dispersion pour chaque variable explicative


```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
ax1.scatter(x1, y)
ax1.set_xlabel('X1')
ax1.set_ylabel('Y')
ax2.scatter(x2, y)
ax2.set_xlabel('X2')
ax2.set_ylabel('Y')
plt.show()
```
